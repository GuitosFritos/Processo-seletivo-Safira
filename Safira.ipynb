# Case middle - Safira Energia

Inicialmente começaremos importando as bibliotecas que usaremos para a resolução do case.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

A partir do arquivo disponibilizado, separei em outras planilhas com o dados mensais. Sendo assim, os importaremos abaixo para realização do case:

```python
dados_marco=pd.read_excel('Março.xlsx', index_col='Cenário')
dados_abril=pd.read_excel('Abril.xlsx', index_col='Cenário')
dados_maio=pd.read_excel('Maio.xlsx', index_col='Cenário')
dados_junho=pd.read_excel('Junho.xlsx', index_col='Cenário')
dados_julho=pd.read_excel('Julho.xlsx', index_col='Cenário')
dados_agosto=pd.read_excel('Agosto.xlsx', index_col='Cenário')
```

Após a anexação de todos os dados no Jupyter, começaremos a analisar cada mês individualmente.

# Março

Iniciaremos verificando os 5 primeiros dados, para verificação de integridade.

```python
dados_marco.head()
```

Faremos uma descrição estatística dos dados:

```python
dados_marco.describe()
```

Agora iremos demonstrar a partir de gráficos o comportamento dos dados para identificarmos uma tendência e retirar os outliers. Usaremos 3 tipos de gráficos, o de pontos, histograma e boxplot, todos servem para encontrar dados fora do padrão, entretanto, cada um com a sua particularidade analítica.
**Outliers significa dados distantes da tendência.**

```python
plt.scatter(dados_marco.loc[:,'Vazão'],dados_marco.loc[:,'Preço'])
plt.title('Março')
plt.xlabel('Vazão')
plt.ylabel('Preço')
plt.show()
```

```python
sns.boxplot(x=dados_marco['Preço'])
```

```python
plt.hist(dados_marco['Preço'])
plt.show()
```

Quanto de gráfico de pontos, como o boxplot, é perceptível que dados com o preço menores que 40 estão fora da tendência esperada. Por isso, os excluiremos a partir do programa abaixo:

```python
c=1
for i in dados_marco['Preço']:
    if i<40:
        dados_marco=dados_marco.drop(c)
    c+=1
```

Agora veremos os gráficos com os dados atualizados:

```python
plt.scatter(dados_marco['Vazão'],dados_marco['Preço'])
plt.title('Março atualizado')
plt.xlabel('Vazão')
plt.ylabel('Preço')
plt.show()
```

# Abril

Assim como no mês anterior continuaremos a análise seguindo a mesma estrutura e ordens de análise.

Verificação de integridade:

```python
dados_abril.head()
```

Dados estatísticos:

```python
dados_abril.describe()
```

Assim como na análise anterior iremos identificar os dados fora da tendência para excluí-los.

```python
plt.scatter(dados_abril['Vazão'],dados_abril['Preço'])
plt.title('Abril')
plt.xlabel('Vazão')
plt.ylabel('Preço')
plt.show()
```

```python
sns.boxplot(x=dados_abril['Preço'])
```

```python
plt.hist(dados_abril['Preço'])
plt.show()
```

A partir do gráfico de pontos é possível visualizar que alguns dados distoam do padrão, são os de vazão menores que 1 e preços menores que 25, por isso os excluiremos.

```python
c=1
for i in dados_abril['Preço']:
    if i<25 and dados_abril['Vazão'][c]<1:
        dados_abril=dados_abril.drop(c)
    c+=1
```

Agora demonstraremos o gráfico com os dados atualizados.

```python
plt.scatter(dados_abril['Vazão'],dados_abril['Preço'])
plt.title('Abril atualizado')
plt.xlabel('Vazão')
plt.ylabel('Preço')
plt.show()
```

```python
plt.scatter(dados_abril['Vazão'],dados_abril['Preço'],label="Abril")
plt.scatter(dados_marco['Vazão'],dados_marco['Preço'],label="Março")
plt.xlabel('Vazão')
plt.ylabel('Preço')
plt.title('Comparação entre Março x Abril')
plt.legend()
plt.show()
```

Em relação a comparação dos preços dos meses, eles são parecedidos, entretanto, a extensão do valor de abril é maior, pois engloba o intervalo 0 até próximo do 200, enquanto março vai de 40 a cerca de 160.

# Junho

Como de praxe, lá vamos nós realizar as rotinas padrões. Primeiramente verificando a integração.

```python
dados_junho.head()
```

Segue os dados estatísticos do mês de junho:

```python
dados_junho.describe()
```

Faremos as análises de gráficos padrões para identificarmos dados que fujam da tendência.

```python
plt.scatter(dados_junho['Vazão'],dados_junho['Preço'])
plt.title('Junho')
plt.xlabel('Vazão')
plt.ylabel('Preço')
plt.show()
```

```python
sns.boxplot(x=dados_junho['Preço'])
```

```python
plt.hist(dados_junho['Preço'])
plt.show()
```

Ao analisarmos os três gráficos percebemos dois problemas, o primeiro são os preços acima de cerca de 225 que estão muito acima da tendência. Além disso, como no mês anterior temos inúmeras vazões para um mesmo preço 0, o qual contraria o padrão de distribuição do gráfico. Sendo assim, os excluiremos.

```python
c=1
for i in dados_junho['Preço']:
    if i==0 or i>225:
        dados_junho=dados_junho.drop(c)
    c+=1
```

Segue o gráfico atualizado:

```python
plt.scatter(dados_junho['Vazão'],dados_junho['Preço'])
plt.title('Junho atualizado')
plt.xlabel('Vazão')
plt.ylabel('Preço')
plt.show()
```

Agora faremos a análise em relação aos meses anteriores. Março, abril e maio são extremante semelhantes, praticamente iguais, por isso usaremos como comparação o mês de maio.

```python
plt.scatter(dados_junho['Vazão'],dados_junho['Preço'],label="Junho")
plt.scatter(dados_maio['Vazão'],dados_maio['Preço'],label="Maio")
plt.xlabel('Vazão')
plt.ylabel('Preço')
plt.title('Comparação entre Maio x Junho')
plt.legend()
plt.show()
```

Os gráficos continuam semelhantes, entretanto, alguns dados de junho estão saindo do padrão que existia em maio, pois há um aumento em pontos de vazões mais altas.

Comparando os dados entre maio e julho é possível perceber que o gráfico de julho está indicando um aumento da vazão. Vemos que em maio não tínhamos dados para vazões maiores que 1,4 e agora começamos a ter bastante pontos. Indicando um aumento da vazão média e uma possível diminuição no preço

# Cenário semestral e histórico:

Para analisarmos o semestre usaremos as médias de cada mês para compararmos as evoluções entre preço e vazão.

Primeiro criaremos um dicionário para futuramente transformá-lo em um dataframe.

```python
analise={'Meses':['Março','Abril','Maio','Junho','Julho','Agosto'],'Vazão':[np.mean(dados_marco['Vazão']),np.mean(dados_abril['Vazão']),np.mean(dados_maio['Vazão']),np.mean(dados_junho['Vazão']),np.mean(dados_julho['Vazão']),np.mean(dados_agosto['Vazão'])],'Preço':[np.mean(dados_marco['Preço']),np.mean(dados_abril['Preço']),np.mean(dados_maio['Preço']),np.mean(dados_junho['Preço']),np.mean(dados_julho['Preço']),np.mean(dados_agosto['Preço'])]}
```

```python
analise=pd.DataFrame(data=analise)
```

Está pronto o nosso banco de dados:

```python
analise.head()
```

Agora faremos análise de como as coisas se mostram no semestre:

Primeiramente faremos da vazão média por mês.

```python
plt.plot(analise['Meses'],analise['Vazão'])
plt.ylabel('Vazão')
plt.xlabel('Meses')
plt.title('Vazão médio por mês')
```

Como podemos ver a média da vazão sobe em abril, em relação a março, e em maio e junho a vazão cai. A partir de julho e agosto a vazão sobe consideravelmente. As alterações na vazão provavelmente afetará o preço médio.

```python
plt.plot(analise['Meses'],analise['Preço'])
plt.ylabel('Preço')
plt.xlabel('Meses')
plt.title('Preço médio por mês')
```

Como é ilustrado no gráfico, vemos que durante o semestre houve uma diminuição no preço médio em todos os meses, o que nos leva a imaginar que não é apenas a vazão que influencia no preço, mas também o mês. Entretanto, os meses os quais a vazão subiu, o preço caiu com a mesma velocidade.
